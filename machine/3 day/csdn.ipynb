```python
def my_salary_tax():
    x = input("输入不小于0的数 ")
    x = int(x)
    y = x -5000
    if y>0 and y<=31000:
        y = y*3/100
        return y
    if y>31000 and y<=139000:
        y = y*10/100 - 2520
        return y
    if y>139000:
        y = y*20/100 - 16920
        return y
    else:
        return 0
```

```python
def salary_tax():
    x = input("输入不小于0的数 ")
    x = int(x)
    if x>5000:
        if x<=36000:
            y = (x-5000)*3/100
        if x>36000 and x<=144000:
            y = (x-5000)*10/100 - 2520
        if x>144000:
            y = (x-5000)*20/100 - 16920
        return y
    else:
        return 0
```

```python
from sklearn import preprocessing
num_variable=['blog_num','view_num','fans_num','likes_num','comments_num','collections_num']
df2=pd.concat([df1[['blog_name','code_time']],
              pd.DataFrame(preprocessing.scale(df1[num_variable]),columns=num_variable)],axis=1)
df2
```

```python
import numpy.linalg as nlg
C=df2.corr()
C
```

```python
eig_value,eig_vector=nlg.eig(C)
```

```python
eig=pd.DataFrame() #利用变量名和特征值建立一个数据框
eig
```

```python
eig['names']=df2.columns[2:]#列名
eig['eig_value']=eig_value#特征值
eig
```

```python
# 确定公共因子个数
for k in range(1,7):
    var_cum = eig['eig_value'][:k].sum()/eig['eig_value'].sum()
    print('累计方差贡献率为{}'.format(var_cum))
    if var_cum>=0.8:
        print('确定公共因子数为{}'.format(k))
        break
```

```python
from factor_analyzer import FactorAnalyzer,Rotator
data1 = df2[num_variable]  #提取数值型变量
fa = FactorAnalyzer(3,rotation=None)
fa.fit(data1)
print('未旋转前因子载荷矩阵:\n',fa.loadings_)

rotator = Rotator()
print("按正交旋转后的因子载荷矩阵:\n", rotator.fit_transform(fa.loadings_))
```

```python
fa1 = FactorAnalyzer(3,rotation='varimax')  #使用正交旋转
fa1.fit(data1)
fa1.get_factor_variance()
```

```python
def score(factors):
    return sum(factors*fa1.get_factor_variance()[1])
scores = []
for i in range(len(fa1.transform(data1))):
    new = score(fa1.transform(data1)[i])
    scores.append(new)
df2['scores']=scores
```

```python
import pandas as pd
import numpy as np
import re
import json
import requests
from gopup.economic import cons
```

```python
def get_gdp_year():
    """
    获取年度国内生产总值数据
    --------
    DataFrame
    year :统计年度
    gdp :国内生产总值(亿元)
    pc_gdp :人均国内生产总值(元)
    gnp :国民生产总值(亿元)
    pi :第一产业(亿元)
    si :第二产业(亿元)
    industry :工业(亿元)
    cons_industry :建筑业(亿元)
    ti :第三产业(亿元)
    trans_industry :交通运输仓储邮电通信业(亿元)
    lbdy :批发零售贸易及餐饮业(亿元)
    """
    rdint = cons.random()
    r = requests.get(cons.MACRO_URL%(cons.P_TYPE['http'], cons.DOMAINS['sina'],
                                    rdint, cons.MACRO_TYPE[0], 0, 70,
                                    rdint))
    text = r.text
    #text = text.decode('gbk') if ct.PY3 else text
    regSym = re.compile(r'\,count:(.*?)\}')
    datastr = regSym.findall(text)
    datastr = datastr[0]
    datastr = datastr.split('data:')[1]
    datastr = datastr.replace('"', '').replace('null', '0')
    js = json.loads(datastr)
    df = pd.DataFrame(js, columns=cons.GDP_YEAR_COLS)
    df[df==0] = np.NaN
    return df
```

```python
df = get_gdp_year()
print(df)
```

```python
def get_cpi():
    """
        获取居民消费价格指数数据
    --------
    DataFrame
        month :统计月份
        cpi :价格指数
    """
    rdint = cons.random()
    r = requests.get(cons.MACRO_URL%(cons.P_TYPE['http'], cons.DOMAINS['sina'],
                                    rdint, cons.MACRO_TYPE[1], 0, 600,
                                    rdint))
    text = r.text
    # # text = text.decode('gbk') if ct.PY3 else text
    regSym = re.compile(r'\,count:(.*?)\}')
    datastr = regSym.findall(text)
    datastr = datastr[0]
    datastr = datastr.split('data:')[1]
    js = json.loads(datastr)
    df = pd.DataFrame(js, columns=cons.CPI_COLS)
    df['cpi'] = df['cpi'].astype(float)
    return df
```

```python
df = get_cpi()
print(df)
```
