```python
def missing_cal(df):
    """
    df :数据集
    return：每个变量的缺失率
    """
    missing_pct = df.isnull().sum()/df.shape[0]
    missing_df = pd.DataFrame(missing_pct).reset_index()
    missing_df = missing_df.rename(columns={'index':'col', 0:'missing_pct'})
    missing_df = missing_df.sort_values('missing_pct',ascending=False).reset_index(drop=True)
    return missing_df
missing_cal(data)
```
